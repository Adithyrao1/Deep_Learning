```python
X = df.iloc[:,:-1]
y = df['Chance of Admit ']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
modell = Sequential()
```

```python
modell.add(Dense(7,input_dim=7,activation='relu'))
modell.add(Dense(7,activation='relu'))
modell.add(Dense(1,activation='linear'))
```

```python
modell.summary()
```

```python
modell.compile(loss='mean_squared_error',optimizer='Adam')
hist = modell.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
```

```python
y_pred = modell.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
